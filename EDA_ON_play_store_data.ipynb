```python
def fill_str_to_avg(a):
  sum=0
  count=0

  for i in range(len(a)):
    if type(a[i])==float:

      sum += a[i]

      count +=1





  for i in range(len(a)):


    if type(a[i])==str :

      a[i]=(sum)/count

  return a

```
